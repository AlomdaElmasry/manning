```python
def count_word(word):
    count_spam=0
    count_ham=0
    for email in emails:
        if word in email['word_count']:
            if email['spam']:
                count_spam += 1
            else:
                count_ham += 1
    return count_spam, count_ham
print count_word('money')
print count_word('lose')
print count_word('weight')
```

```python
def probability_spam(word):
    print word
    return 1.0*count_word(word)[0]/(count_word(word)[0]+count_word(word)[1])
probability_spam('money')
```

```python
import sklearn
import numpy
```

```python
y = numpy.array(emails['spam'])
```

```python
x = turicreate.SFrame({'word_count': emails['word_count']})
```

```python
all_word_appearances = {}
for email in emails:
    for word in email['word_count']:
        if word in all_word_appearances:
            all_word_appearances[word] += email['word_count'][word]
        else:
            all_word_appearances[word] = email['word_count'][word]            
```

```python
word_appearances = {}
for word in all_word_appearances:
    if all_word_appearances[word] >= 10 and word != 'subject':
        word_appearances[word] = all_word_appearances[word]
```

```python
def word_in_email(email, word):
    return word in email['word_count']
```

```python
count = 0
columns = []
for word in word_appearances:
    count += 1
    if count % 100 == 0:
        print count
    #print word
    #print emails.apply(lambda x: word_in_email(x, word))
    #emails[word] = emails.apply(lambda x: word_in_email(x, word))
    columns.append(emails.apply(lambda x: word_in_email(x, word)))
```

```python
emails_new = numpy.transpose(numpy.array(columns))
emails_new
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
model = gnb.fit(emails_new, y)
```

```python
y_pred = model.predict(emails_new)
```
